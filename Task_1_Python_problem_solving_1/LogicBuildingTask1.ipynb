## Task 2: Pass / Fail Analyzer

**Problem Statement:**  
Analyze student results and determine pass/fail counts.

```python
marks=[45, 78, 90, 33, 60]
PassCount=0
FailCount=0
TotalStudents=len(marks)

for i in marks:
    if i>=50:
        PassCount+=1
    else:
        FailCount+=1

print(f"Total Students:{TotalStudents}")
print(f"Students Passed:{PassCount}")
print(f"Students Failed:{FailCount}")
```

## Task 3: Simple Data Cleaner

**Problem Statement:**  
Clean and standardize user names.

```python
names=[" Alice ","bob"," CHARLIE "]
CleanedNames = []
for i in names:
    CleanedName = i.strip().lower() #strip method to remove extra spaces and lower method to make all characters lower
    CleanedNames.append(CleanedName)

print("Original Names:",names)
print("Cleaned Names:",CleanedNames)
```

## Task 4: Message Length Analyzer

**Problem Statement:**  
Analyze message lengths and flag long messages.

```python
messages=["Hi", "Welcome to the platform", "OK"]

for i in messages:
    length = len(i)
    print(f"Message:\"{i}\" (Length={length})")
    
    if length > 10:
        print(f"Message:\"{i}\" (This message is longer than 10 characters)")
```
