## 1 Smart Light Controller
**Problem Statement:**
Create a class to control a smart light that can be turned ON or OFF and display its status.

**Sample Input:**

Light Name: Bedroom Light

Action: ON

**Expected Output:**

Bedroom Light is ON

```python
# Smart light class definition
class SmartLight:

    def __init__(self,name):
        self.name=name
        self.status="OFF"

    # method to control light(ON or OFF)
    def control(self,action):
        action=action.strip().upper()
        if action=="ON":
            self.status="ON"
        elif action=="OFF":
            self.status="OFF"

    # method to display status
    def DisplayStatus(self):
        print(f"{self.name} is {self.status}")


# Taking user input
LightName=input("Light Name: ")
action=input("Action (ON/OFF): ")

# Object creation
light=SmartLight(lightName)
light.control(action)

# Display result
light.DisplayStatus()
```

## 2. Employee ID Card System
**Problem Statement:**
Create a class that stores employee name, ID, and department and displays ID card details.

**Sample Input:**

Employee Name: Rahul

Employee ID: EMP102


Department: IT

**Expected Output:**

Employee ID Card

Name: Rahul

ID: EMP102

Department: IT

```python
# Employee Class definition
class Employee:
    
    def __init__(self,name,empID,department):
        self.name=name
        self.empID=empID
        self.department=department

    # method to display ID card
    def DisplayIDCard(self):
        print("\nEmployee ID Card")
        print("Name:",self.name)
        print("ID:",self.empID)
        print("Department:",self.department)


# Taking user input
name=input("Employee Name: ")
empID=input("Employee ID: ")
department=input("Department: ")

# Object creation
emp=Employee(name,empID,department)

# Display result
emp.DisplayIDCard()
```

## 3. Mobile Contact Record
**Problem Statement:**
Create a class to store contact name and phone number and display contact information.

**Sample Input:**

Contact Name: Anita

Phone Number: 9876543210

**Expected Output:**

Contact Saved

Name: Anita

Phone: 9876543210

```python
# Contact class definition
class Contact:

    def __init__(self,name,PhoneNumer):
        self.name=name
        self.PhoneNumber=PhoneNumber

    # method to display contact info
    def DisplayContact(self):
        print("\nContact Saved")
        print("Name:",self.name)
        print("Phone:",self.PhoneNumber)


# Taking user input
name=input("Contact Name: ")
PhoneNumber=input("Phone Number: ")

# Object creation
contact=Contact(name,PhoneNumber)

# Display result
contact.DisplayContact()
```
