## 1 Social Media – Post Engagement Analyzer
**Problem Statement:**
Create a function to analyze engagement on a social media post.

**Rules:**
- Calculate total likes
- If Total Likes ≥ 1000 → Viral Post
- Otherwise → Normal Engagement

**Expected Output Format:**

Total Likes: 1050

Post Status: Viral Post

```python
# Function to analyze engagement
def AnalyzeEngagement(likes):
    # Calculate total likes
    TotalLikes=0
    for i in likes:
        TotalLikes+=i

    # Check post status
    if TotalLikes>=1000:
        status="Viral Post"
    else:
        status="Normal Engagement"

    # Display output
    print(f"Total Likes: {TotalLikes}")
    print(f"Post Status: {status}")


# Take input from uer 
likes=list(map(int,input("Enter likes separated by comma: ").strip().split(',')))
# Function call
AnalyzeEngagement(likes)
```

## 3. Agriculture – Rainfall Adequacy Checker
**Problem Statement:**
Create a function that analyzes seasonal rainfall data.

**Rules:**
- Calculate average rainfall
- Average ≥ required level → Adequate Rainfall
- Otherwise → Inadequate Rainfall

**Expected Output Format:**
    
Average Rainfall: 72

Rainfall Status: Adequate Rainfall

```python
# Function to calculate rainfall status
def RainfallChecker(RainfallList,RequiredLevel):
    total=0
    # Loop to calculate total rainfall
    for i in RainfallList:
        total+=i
    # Calculate average
    average=total/len(RainfallList)

    # Check adequacy
    if average>=RequiredLevel:
        status="Adequate Rainfall"
    else:
        status="Inadequate Rainfall"

    # Output
    print(f"Average Rainfall: {int(average)}")
    print(f"Rainfall Status: {status}")


# Take input from usr
rainfall=list(map(int, input("Enter rainfall values separated by comma: ").strip().split(',')))
required=int(input("Enter required rainfall level: ").strip())

# Function call
RainfallChecker(rainfall,required)
```

## 4. Social Media – Duplicate Account Detection
**Problem Statement:**
Create a function to detect duplicate usernames in a system.
    
**Rules:**
- If duplicates exist → Flag accounts
- Else → Accounts are unique

**Expected Output Format:**

Duplicate Accounts Found: Yes

```python
# Function to detetct duplicates
def DetectDuplicates(usernames):
    # Using set method to check duplicates
    if len(usernames)!=len(set(usernames)):
        result="Yes"
    else:
        result="No"

    # Output
    print(f"Duplicate Accounts Found: {result}")

# Take input from user
usernames=input("Enter usernames separated by comma: ").strip().split(',')
usernames=[name.strip() for name in usernames]

# Function call
DetectDuplicates(usernames)
```

## 6. Agriculture – Premium Crop Price Filter
**Problem Statement:**
Create a function that filters crop prices above a premium threshold.

**Rules:**
- Price > ₹2000 → Premium Crop
- Else → Regular Crop

**Expected Output Format:**

Premium Crops: [2500, 3200]

```python
# Function to filter premium crop
def PremiumCropFilter(prices):
    # List filtering using loop
    premium=[]
    for i in prices:
        if i>2000:
            premium.append(i)

    # Output
    print(f"Premium Crops: {premium}")


# Take user input
prices=list(map(int,input("Enter crop prices separated by comma: ").strip().split(',')))

# Function call
PremiumCropFilter(prices)
```

## 9. E-Learning – Student Attendance Eligibility System
**Problem Statement:**
Create a function to calculate attendance percentage and check exam eligibility.

**Rules:**
- Attendance ≥ 75% → Eligible
- Attendance < 75% → Not Eligible

**Expected Output Format:**

Attendance Percentage: 80.0

Exam Eligibility: Eligible

```python
# Function to check attandance
def AttendanceChecker(attendance):
    # Logic to calculate percentage
    total=0
    for i in attendance:
        total+=i
    percentage=(total/len(attendance))*100    # Calculate percentage

    # Eligibility check
    if percentage>=75:
        status="Eligible"
    else:
        status="Not Eligible"

    # Output
    print(f"Attendance Percentage: {percentage}")
    print(f"Exam Eligibility: {status}")


# TAke input from usr (P=1, A=0 style)
attendance=list(map(int,input("Enter attendance (1 for Present, 0 for Absent) separated by comma: ").strip().split(',')))

# Function call
AttendanceChecker(attendance)
```

## 10. Smart Electricity Bill Analyzer
**Problem Statement:**
You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.

**Billing Rules:**
- First 100 units → ₹3 per unit
- Next 100 units (101–200) → ₹5 per unit
- Above 200 units → ₹7 per unit

**Usage Classification:**
- Total bill < ₹500 → Low Usage
- ₹500 – ₹1500 → Moderate Usage
- ₹1500 → High Usage

```python
# Function to calculate the electricity bill & usage status
def CalculateBill(units):
    bill=0
    # Slab calculation according to billing rules
    if units<=100:
        bill=units*3
    elif units<=200:
        bill=(100*3)+((units-100)*5)
    else:
        bill=(100*3)+(100*5)+((units-200)*7)

    # Usage classification
    if bill<500:
        status="Low Usage"
    elif bill<=1500:
        status="Moderate Usage"
    else:
        status="High Usage"

    return bill,status


# Take user input
units=int(input("Enter units consumed: ").strip())

# Function call
TotalBill,UsageStatus=CalculateBill(units)

# Display Output
print(f"Total Bill: ₹{TotalBill}")
print(f"Usage Status: {UsageStatus}")
```
