## 1. Mobile Recharge Validation System
**Problem Statement**
Create a function to validate a mobile recharge request.

**Rules:**
- Recharge amount must be â‰¥ â‚¹50
- Valid plans: 199, 299, 399, 599
- User should be allowed to retry if invalid

```python
# Function to validate recharge
def ValidateRecharge(amount,ValidPlans):
    if amount<50:
        print("To Recharge amount must be at least â‚¹50.")
        print(f"Available Plans:{ValidPlans}")
        return False
    elif amount not in ValidPlans:
        print(f"{amount} plan isn't available right now")
        print(f"Available Plans:{ValidPlans}")
        return False
    return True

# To show user available plans
ValidPlans=[199,299,399,599]
print(f"Available Plans:{ValidPlans}")

# To ask user, mobile number
MobileNumber=input("Enter mobile number: ").strip()

# While loop for retry
while True:
    amount=int(input("Enter recharge plan amount: â‚¹"))
    print("\n")
    if ValidateRecharge(amount,ValidPlans):
        print(f"Recharge successful for {MobileNumber} with plan â‚¹{amount}")
        break
    else:
        print("Please try again.\n")
        print(f"Available Plans:{ValidPlans}")
```

## 2. Inventory Reorder Alert System
**Problem Statement:**
Given a dictionary of products and stock quantities:
- Print **Reorder Alert** if stock < 15
- Print **Stock OK** otherwise

```python
# Function to check inventory
def CheckInventory(inventory):
    print("\nInventory Status:")
    for product,stock in inventory.items():
        if stock<15:
            print(f"{product}=Reorder Alert")
        else:
            print(f"{product}=Stock OK")

# Input from user
inventory={}
n = int(input("Enter number of products: "))
for i in range(n):
    product,quantity=input(f"Enter Product data {i+1} (Produt:Quantity): ").split(":")
    inventory[product.strip()]=int(quantity.strip())

# Calling function
CheckInventory(inventory)
```

## 3. Student Result Processing System
**Problem Statement:**
Create a function that:
- Accepts a list of student marks
- Calculates average marks
- Prints **Pass** if average â‰¥ 50 else **Fail**

```python
# Function to check student result
def CheckResult(marks):
    if not marks:
        print("No marks provided.")
        return
        
    # ðŸ” Loop to calculate sum and length
    total=0
    count=0
    for mark in marks:
        total+=mark
        count+=1

    average=total/count
    print(f"\nAverage Marks: {average:.2f}")

    if average>=50:
        print("Result: Pass")
    else:
        print("Result: Fail")

# Take input from user
marks=list(map(int,input("Enter marks of the student: ").strip().split(',')))
CheckResult(marks)
```

## 4. Cab Fare Estimator with Retry Option
**Problem Statement**
Create a cab fare calculator:
-Base fare = â‚¹50
- â‚¹12 per km
- Peak hour â†’ 25% extra
- Allow retry using while loop

```python
# Function to calculate fare
def CalculateFare(distance,IsPeak):
    BaseFare=50
    PerKmRate=12
    fare=BaseFare+(distance*PerKmRate)

    if IsPeak == 'y':
        fare*=1.25

    return round(fare)

while True:
    # Get valid distance
    while True:
        distance=input("Enter distance travelled (in km): ").strip()

        # Validating distance input
        if not distance.replace('.','',1).isdigit():
            print("Please enter a valid numeric distance.\n")
            continue

        distance=float(distance)
        break

    # Step 2: Get valid peak input
    while True:
        peak=input("Is it peak hour? (y/n): ").strip().lower()

        # Validating peak input
        if peak not in ['y', 'n']:
            print("Please enter 'y' or 'n' only.\n")
            continue
        break

    # Step 3: Confirmation
    confirm=input(f"Are the distance({distance})Km and peak({peak}) values correct? (y/n): ").strip().lower()

    if confirm=='y':
        print(f"Total Cab Fare: â‚¹{CalculateFare(distance,peak)}")
        break
    else:
        print("Re-enter the details.\n")
```

## 5. Employee Attendance Eligibility Checker
**Problem Statement:**
Create a function that:
- Accepts attendance list ("P", "A")
- Calculates attendance percentage
- Returns Eligible if â‰¥ 75% else Not Eligible

```python
# Function to check eligibility
def CheckEligibility(attendanceList):
    PresentCount=0
    TotalDays=0

    # Loop to count present days
    for status in attendanceList:
        TotalDays+=1
        if status.upper()=="P":
            PresentCount+=1

    AttendancePercentage=(PresentCount/TotalDays)*100

    if AttendancePercentage>=75:
        return "Eligible"
    else:
        return "Not Eligible"


# Take input from user
AttendanceInput=input("Enter attendance (P/A separated by comma): ").strip().split(',')
AttendanceList=[x.strip().upper() for x in AttendanceInput]

# Call function and print result
result=CheckEligibility(AttendanceList)
print(f"Attendance Percentage: {(AttendanceList.count('P')/len(AttendanceList))*100:.2f}%")
print(f"Status: {result}")
```

## 6. Password Strength Checker
**Problem Statement:**
Create a function to check password strength.

**Rules:**
- Minimum length = 8
- Must contain at least one digit
- Must contain at least one special character (@#$)

```python
# Function to check password strength
def CheckPasswordStrength(password):
    HasDigit=False
    HasSpecial=False
    PasswordLength=0
    # Loop over string
    for i in password:
        PasswordLength+=1
        if i.isdigit():
            HasDigit=True
        if not i.isalnum():
            HasSpecial=True

    # Conditions
    if PasswordLength< 8:
        return "Weak Password"
    if not HasDigit:
        return "Weak Password"
    if not HasSpecial:
        return "Weak Password"

    return "Strong Password"


# Take input from user
password = input("Enter password: ").strip()
print(f"Password Strength: {CheckPasswordStrength(password)}")    # Call function
```
